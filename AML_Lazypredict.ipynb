```python
# Read in data and packages
from lazypredict.Supervised import LazyClassifier
from lazypredict.Supervised import LazyRegressor
from sklearn.model_selection import train_test_split

import pandas as pd
import numpy as np

df = pd.read_csv('ML1_data_adj_v3.csv')
```

### LAZYPREDICT REGRESSION

```python
df1 = df.drop(columns=['Date', 'ticker', 'Fwd_Return_3M', 'Fwd_Return_3M_Class'])
df1.describe()
```

```python
X = df1.drop(columns=['Fwd_Best_EPS_6M'])
X.describe()
```

```python
y = df1[['Fwd_Best_EPS_6M']]
y.describe()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=.3,random_state =123)
```

```python
reg = LazyRegressor(verbose=0, ignore_warnings=False, custom_metric=None)
models, predictions = reg.fit(X_train, X_test, y_train, y_test)

print(models)
```

### LAZYPREDICT CLASSIFICATION

```python
df2 = df.drop(columns=['Date', 'ticker', 'Fwd_Return_3M', 'Fwd_Best_EPS_6M'])
df2.describe()
```

```python
X2 = df2.drop(columns=['Fwd_Return_3M_Class'])
X2.describe()
```

```python
y2 = df2[['Fwd_Return_3M_Class']]
y2.describe()
```

```python
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2,test_size=.3,random_state =123)
```

```python
clf = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None)
models,predictions = clf.fit(X2_train, X2_test, y2_train, y2_test)
print(models)
```
